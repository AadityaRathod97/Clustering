```python
import pandas as pd 
import numpy  as np
import matplotlib.pyplot as plt

Univ = pd.read_csv("C:\\Users\\LENOVO\\Desktop\\ExcelR_datascience\\Python Codes\\Clustering\\Universities.csv")
```

```python
Univ
```

```python
Univ.describe()
```

```python
def norm_func(i):
    x = (i-i.min())	/	(i.max()	-	i.min())
    return (x)
```

```python
df_norm = norm_func(Univ.iloc[:,1:])
df_norm
```

```python
type(df_norm)
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
```

```python
z = linkage(df_norm, method="complete",metric="euclidean")
```

```python
z
```

```python
plt.figure(figsize=(15, 5))
```

```python
plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(
    z,
    leaf_rotation=0.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
help(linkage)
```

```python
# Now applying AgglomerativeClustering choosing 3 as clusters from the dendrogram
from	sklearn.cluster	import	AgglomerativeClustering 
h_complete	=	AgglomerativeClustering(n_clusters=3,	linkage='complete',affinity = "euclidean").fit(df_norm) 
```

```python
cluster_labels=pd.Series(h_complete.labels_)
Univ['clust']=cluster_labels # creating a  new column and assigning it to new column 
Univ = Univ.iloc[:,[7,0,1,2,3,4,5,6]]
Univ.head()
```

```python
# getting aggregate mean of each cluster
Univ.iloc[:,2:].groupby(Univ.clust).mean()
```
